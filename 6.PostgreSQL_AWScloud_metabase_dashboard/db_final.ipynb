```python
# 1. Get the names and the quantities in stock for each product.

engine.execute('CREATE TABLE q_1 AS (SELECT "productName","unitsInStock" FROM products);')
df_1 = pd.read_sql('q_1', engine, index_col=0)
df_1.head()
```

```python
# 2. Get a list of current products (Product ID and name)

engine.execute('CREATE TABLE q_2 AS (SELECT "productID","productName" FROM products);')
df_2 = pd.read_sql('q_2', engine, index_col='productID')
df_2.head()
```

```python
# 3. Get a list of the most and least expensive products (name and unit price)

engine.execute('CREATE TABLE q_3 AS (select "productName", "unitPrice" from products where "unitPrice" = (select min("unitPrice") from products) or "unitPrice" = (select max("unitPrice") from products));')
df_3 = pd.read_sql('q_3', engine, index_col='productName')
df_3
```

```python
# 4. Get products that cost less than $20

engine.execute('CREATE TABLE q_4 AS (select "productName", "unitPrice" from products where "unitPrice" < 20 order by "unitPrice");')
df_4 = pd.read_sql('q_4', engine, index_col='productName')
df_4.head()
```

```python
# 5. Get products that cost between $15 and $25.

engine.execute('CREATE TABLE q_5 AS (select "productName", "unitPrice" from products where "unitPrice" >14 and "unitPrice" < 26 order by "unitPrice");')
df_5 = pd.read_sql('q_5', engine, index_col='productName')
df_5.head()
```

```python
# 6. Get products above average price.

engine.execute('CREATE TABLE q_6 AS (select "productName", "unitPrice" from products where "unitPrice" > (select AVG("unitPrice") from products) order by "unitPrice");')
df_6 = pd.read_sql('q_6', engine, index_col='productName')
df_6.head()
```

```python
# 7. Find the 10 most expensive products.

engine.execute('CREATE TABLE q_7 AS (select "productName", "unitPrice" from products order by "unitPrice" desc limit 10);')
df_7 = pd.read_sql('q_7', engine, index_col='productName')
df_7
```

```python
# 8. Get a list of discontinued products (Product ID and name)

engine.execute('CREATE TABLE q_8 AS (select "productID","productName","discontinued" from products where "discontinued" = 1);')
df_8 = pd.read_sql('q_8', engine, index_col='productID')
df_8
```

```python
# 9. Count current and discontinued products.

engine.execute('CREATE TABLE q_9 AS (select min(discontinued) as discontinued, count(discontinued) from products where  discontinued = 0 union select max(discontinued) as discontinued, count(discontinued) from products where  discontinued = 1);')
df_9 = pd.read_sql('q_9', engine, index_col='discontinued')
df_9
```

```python
# 10. Find products with less units in stock than the quantity on order.
# sorted by difference btw on stock and on order

engine.execute('CREATE TABLE q_10 AS (select "productID","productName","unitsInStock","unitsOnOrder" from products where "unitsInStock" < "unitsOnOrder" order by "unitsInStock" - "unitsOnOrder");')
df_10 = pd.read_sql('q_10', engine, index_col='productID')
df_10.head()
```

```python
# 11. Find the customer who had the highest order amount

engine.execute('CREATE TABLE q_11 AS (select o."customerID",c."companyName", o."orderID",od."quantity",od."unitPrice" from order_details as od inner join orders as o on od."orderID"=o."orderID" inner join customers as c on o."customerID"=c."customerID" order by "quantity"*"unitPrice" desc limit 5);')
df_11 = pd.read_sql('q_11', engine, index_col='customerID')
df_11.head()
```

```python
# 12. Get orders for a given employee and the according customer

engine.execute('CREATE TABLE q_12 AS (select o."employeeID",e."firstName", e."lastName", o."customerID", c."companyName", c."country", count(o."customerID") as orders_count from customers as c inner join orders as o on c."customerID"=o."customerID" inner join employees as e on e."employeeID"=o."employeeID" group by o."employeeID", o."customerID", c."companyName", c."country", e."lastName", e."firstName" order by orders_count desc);')
df_12 = pd.read_sql('q_12', engine, index_col='employeeID')
df_12.head()
```
